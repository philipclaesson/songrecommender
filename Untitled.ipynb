```python
# Do all relevant imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.sparse as sps
import time
%matplotlib inline
print("Libs imported.")
```

```python
## Support functions

import matplotlib.pyplot as plt
%matplotlib inline
def plot_coo_matrix(m, size = 0.004):
    if not isinstance(m, sps.coo_matrix):
        m = sps.coo_matrix(m)
    fig = plt.figure()
    ax = fig.add_subplot(111, facecolor='green')
    ax.plot(m.col, m.row, 's', color='black', ms=size)
    ax.set_xlim(0, m.shape[1])
    ax.set_ylim(0, m.shape[0])
    ax.set_aspect('equal')
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.invert_yaxis()
    ax.set_aspect('equal')
    ax.set_xticks([])
    ax.set_yticks([])
    return ax
```

```python
ICM_perfect_7 = sps.load_npz('Saved Matrixes/ICM_perfect_7.npz')


ax = plot_coo_matrix(ICM_perfect_7)
ax.figure.show()

ICM_perfect_7.dtype = np.int32

print(ICM_perfect_7[:,1000])
```
